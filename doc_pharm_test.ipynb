```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data input
doctors = pd.read_csv("c:/Users/Laurens/Documents/IE-BIGDATA/Personal projects/Raphael doctors distance/doctors.csv", encoding='latin-1')
```

```python
pharms = pd.read_csv("c:/Users/Laurens/Documents/IE-BIGDATA/Personal projects/Raphael doctors distance/pharmacies.csv")
```

```python
# drop null values
doc = doctors.dropna()
pharms = pharms.dropna()
```

```python
# distances function
def get_distance(d_lat,d_lon,p_lat,p_lon):
    lat_diff = (p_lat - d_lat)**2
    lon_diff = (p_lon - d_lon)**2
    return np.sqrt(lat_diff + lon_diff)
```

```python
# check that function works
get_distance(doc.loc[0,'lat'],doc.loc[0,'lon'],
            pharms.loc[0,'lat'],pharms.loc[0,'lon'])
```

```python
# empty dataframe to store the closest pharmacy
# for each doctor
min_df = pd.DataFrame(columns = ['doctor','doctor_ind',
                                 'distance','pharma',
                                 'pharma_ind'])

for di in range(len(doc)):
    #print status
    print("working on index", di, "of", len(doc))
    # keep length of min_df to add new rows
    ind = len(min_df)
    
    # doctor
    drow = doc.iloc[di]
    
    # store distances for each doctor
    d_distances = []
    
    # get distance for each doctor di and pharmacy pi
    for pi in range(len(pharms)):
        # pharmacy
        prow = pharms.iloc[pi]
        
        # calculte distance and store distance value
        dist = get_distance(drow['lat'], drow['lon'],
                           prow['lat'], prow['lon'])
        d_distances.append(dist)
    
    # calculate min distance and index of pharmacy with
    # min distance
    min_dist = min(d_distances)
    min_pharm_ind = d_distances.index(min_dist)
    
    # append doc + min distance pharmacy to dataframe
    min_df.loc[ind,:] = [drow['fulladres'],di,
                        min_dist,pharms.loc[min_pharm_ind,'Adresse'],
                        min_pharm_ind]
```
